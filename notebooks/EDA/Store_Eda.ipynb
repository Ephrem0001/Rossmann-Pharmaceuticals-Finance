```python
df_store["CompetitionDistance"] = df_store["CompetitionDistance"].ffill().bfill()
df_store["CompetitionOpenSinceMonth"] = df_store["CompetitionOpenSinceMonth"].ffill().bfill()
df_store["CompetitionOpenSinceYear"] = df_store["CompetitionOpenSinceYear"].ffill().bfill()
```

```python
df_store["Promo2SinceWeek"] = df_store["Promo2SinceWeek"].fillna(df_store["Promo2SinceWeek"].median())  # Imputing with median week
df_store["Promo2SinceYear"] = df_store["Promo2SinceYear"].fillna(df_store["Promo2SinceYear"].median())  # Imputing with median week
```

```python
df_store["PromoInterval"] = df_store["PromoInterval"].ffill().bfill()
```

```python
# Save the cleaned dataset to a CSV file
df_store.to_csv(r"C:\Users\ephre\Documents\Rossmann-Pharmaceuticals-Finance-1\Data\Raw_Data\cleaned_Store_dataset.csv", index=False)
```
